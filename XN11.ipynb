```python
def fn(pname):
  l1=[]
  l1.append(pname)
  bats_man=ball[ball['batsman']==pname]
  run_score=bats_man['batsman_runs'].sum()
  bowls_faced=len(bats_man)-len(bats_man[bats_man['wide_runs']!=0])
  strike_rate=(run_score*100)/bowls_faced
  bowl_er=ball[ball['bowler']==pname]
  run_conceeded=bowl_er['total_runs'].sum()
  bowls_bowled=len(bowl_er)-len(bowl_er[bowl_er['wide_runs']!=0])-len(bowl_er[bowl_er['noball_runs']!=0])
  # economy=run_conceeded/bowls_bowled
  wc=bowl_er['dismissal_kind'].isnull().sum()
  wc=len(bowl_er)-wc
  wc=wc-len(bowl_er[bowl_er['dismissal_kind']=='run out'])
  wickets=wc
  stumpings=len(bowl_er)-wickets
  wk=0;
  if(stumpings!=0):
    wk=1
  out=len(ball[ball['player_dismissed']==pname])
  dht={}
  for key in bats_man['match_id']:
    dht[key]=0;
  mat=len(dht)
  notout=mat-out;
  for key in bowl_er['match_id']:
    dht[key]=0;
  match_played=len(dht)
  average=-1
  if(out!=0):
    average=run_score/out
  over_bowled=bowls_bowled/6;
  # over_bowled=int(over_bowled)
  bowler_average=-1
  economy=-1;
  if(over_bowled!=0):
    economy=run_conceeded/over_bowled
  
  if(wc!=0):
    bowler_average=run_conceeded/wc
  
  l1.append(match_played)
  l1.append(run_score)
  average=round(average,2)
  l1.append(average)
  strike_rate=round(strike_rate,2)
  l1.append(strike_rate)
  l1.append(notout)
  over_bowled=round(over_bowled,2)
  l1.append(over_bowled)
  l1.append(wickets)
  economy=round(economy,2)
  l1.append(economy)
  round(bowler_average,2)
  l1.append(bowler_average)
  #wicketkeeper
  field=ball[ball['fielder']==pname]
  stumpings=len(field[field['dismissal_kind']=='stumped'])
  l1.append(stumpings)
  runout=len(field[field['dismissal_kind']=='run out'])
  l1.append(runout)
  caughts=len(field[field['dismissal_kind']=='caught'])
  l1.append(caughts)
  manofmatch=len(match[match['player_of_match']==pname])
  l1.append(manofmatch)
  return l1
  # run_score=bats_man['bats']
```

```python
season_win=match.groupby('season')['winner'].value_counts()
season_win
```

```python
season_win.head()
```

```python
year = 2008
win_per_season_df = pd.DataFrame(columns=['year', 'team', 'wins'])
for items in season_win.iteritems():    
    if items[0][0]==year:
        # print(items)
        win_series = pd.DataFrame({
            'year': [items[0][0]],
            'team': [items[0][1]],
            'wins': [items[1]]
        })
        win_per_season_df = win_per_season_df.append(win_series)
        year += 1   
```

```python
win_per_season_df
```

```python
sns.barplot('wins', 'team', hue='year', data=win_per_season_df, palette='Paired');
```
